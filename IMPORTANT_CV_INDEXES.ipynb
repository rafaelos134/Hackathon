```python
import pandas as pd
import glob
import os

# Define the path pattern to match all Parquet files in the directory
# Using os.path.join to create a platform-independent path
folder_path = os.path.join('data', 'hackathon_2025_templates')
path_pattern = os.path.join(folder_path, '*.parquet')

# Get a list of all matching Parquet file paths
parquet_files = glob.glob(path_pattern)

if not parquet_files:
    print(f"No .parquet files found in '{folder_path}'")
else:
    print(f"Found {len(parquet_files)} Parquet file(s). Summarizing each...\n")

    # Loop through each found file path
    for file_path in parquet_files:
        print("─" * 80)
        print(f"📊 Summary for: {os.path.basename(file_path)}")
        print("─" * 80)

        try:
            # Read the Parquet file into a pandas DataFrame
            df = pd.read_parquet(file_path)

            # --- Shape ---
            print(f"\nDimensions: {df.shape[0]} rows, {df.shape[1]} columns\n")

            # --- Column Types and Non-Null Counts ---
            print("Column Info:")
            # df.info() prints a lot; this is a more compact version
            for col in df.columns:
                print(f"  - {col:<20} | Type: {str(df[col].dtype):<10} | Non-Nulls: {df[col].notna().sum()}")

            # --- Data Preview ---
            print("\nData Preview (First 5 Rows):")
            print(df.head())

            # --- Numerical Summary ---
            if not df.select_dtypes(include='number').empty:
                print("\nNumerical Data Summary:")
                print(df.describe().round(2))
            
            # --- Categorical Summary ---
            if not df.select_dtypes(include='object').empty:
                print("\nCategorical Data Summary:")
                print(df.describe(include='object'))

            print("\n")

        except Exception as e:
            print(f"\nCould not process file {os.path.basename(file_path)}. Error: {e}\n")
```
